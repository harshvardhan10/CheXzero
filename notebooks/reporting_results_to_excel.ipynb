```python
import json
import pandas as pd
```

```python
with open("/home/woody/iwi5/iwi5362h/CheXzero/vindr_results/vindr_zero_shot_ensemble.json", "r") as f:
    zsclf = json.load(f)
```

```python
# summary table

ckpt_name = zsclf['checkpoint']
cls = zsclf['classification']

summary_rows = []

summary_rows.append({
    "checkpoint": ckpt_name,
    "macro_auc": cls["macro_auc"],
    "micro_auc": cls["micro_auc"],
    "map_at_10": cls["map_at_10"],
})


df_summary = pd.DataFrame(summary_rows)
df_summary
```

```python
per_label_dict = {}

per_auc = cls["per_label_auc"]

# Convert None â†’ NaN for readability
row = {label: (auc if auc is not None else float("nan"))
       for label, auc in per_auc.items()}

per_label_dict[ckpt_name] = row

# Build DataFrame: rows = labels, columns = checkpoints
df_per_label = pd.DataFrame(per_label_dict)
df_per_label
```
